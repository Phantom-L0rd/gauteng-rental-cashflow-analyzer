```python
def calc_bond_payment(P, rate, years):
    r = rate / 12
    n = years * 12
    return npf.pmt(r, n, -P)

def calc_annual_rent(m_rent):
    return m_rent * 12

def calc_vacancy_loss(a_rent, loss_rate):
    return a_rent * loss_rate

def calc_monthly_expenses(levies, rates, maintenance = 0, insurance = 0, v_reserve = 0):
    return levies + rates + maintenance + insurance + v_reserve

def calc_annual_expenses(m_exps):
    return m_exps * 12

def calc_cashflow(m_rent, m_exps, m_mort):
    return m_rent - m_exps - m_mort

def calc_annual_cashflow(cashflow):
    return cashflow * 12

def calc_annual_noi(a_rent, a_exps):
    return a_rent - a_exps

def calc_cap_rate(a_noi, property_value):
    return a_noi / property_value * 100

def calc_roi(a_cashflow, property_price, closing_cost = 70000, reno_costs = 0):
    tot_investments = property_price + closing_cost + reno_costs
    return a_cashflow / tot_investments * 100

def calc_dscr(a_noi, m_mort):
    a_mort = m_mort * 12
    return a_noi / a_mort

def calc_expected_growth_rate(e_appr_rate, infla_rate):
    return e_appr_rate - infla_rate

def calc_break_even(a_exps, m_mort, a_rent):
    a_mort = m_mort * 12
    return (a_exps + a_mort) / a_rent

def calc_investment_score(cashflow, cap_rate, roi, dscr, inflation_adjusted_growth, break_even):
    
    # if cashflow < -(m_mort/2):
    #     cash_i = 0
    # elif cashflow > 3000:
    #     cash_i = 1
    # else:
    #     cash_i = (cashflow + (m_mort/2)) / (3000 + (m_mort/2))

    norm_cashflow = max(min(cashflow / 3000, 1), 0.0)
    norm_cap_rate = max(cap_rate, 0) / 12
    norm_roi = max(roi,0) / 20
    norm_dscr = min(max(dscr,0) / 1.5, 1)
    norm_growth = max((inflation_adjusted_growth + 5) / 10,0)
    norm_break_even = 1 - (break_even / 1.0)

    investment_score = (
        0.25 * norm_cashflow +
        0.20 * norm_cap_rate +
        0.20 * norm_roi +
        0.15 * norm_dscr +
        0.10 * norm_growth +
        0.10 * norm_break_even
    ) * 100

    return  max(min(investment_score, 100), 0)
```

```python
df.loc[:, "monthly_repayments"] = df['sale_price'].apply(lambda x: calc_bond_payment(x,0.1025,20))

df.loc[:, "annual_rent"] = df['pred_rent'].apply(lambda x: calc_annual_rent(x))

df.loc[:, "vacancy_loss"] = df['annual_rent'].apply(lambda x: calc_vacancy_loss(x,0.07))

df.loc[:, "monthly_expenses"] = df.apply(lambda x: calc_monthly_expenses(x['levies'], x['rates_taxes']), axis=1)

df.loc[:, "annual_expenses"] = df["monthly_expenses"].apply(lambda x: calc_annual_expenses(x))

df.loc[:, "cashflow"] = df.apply(lambda x: calc_cashflow(x['pred_rent'], x['monthly_expenses'], x['monthly_repayments']), axis=1)

df.loc[:, "annual_cashflow"] = df['cashflow'].apply(lambda x: calc_annual_cashflow(x))

df.loc[:, "annual_noi"] = df.apply(lambda x: calc_annual_noi(x['annual_rent'], x['annual_expenses']) ,axis=1)

df.loc[:, "cap_rate"] = df.apply(lambda x: calc_cap_rate(x['annual_noi'], x['sale_price']),axis=1)

df.loc[:, "roi"] = df.apply(lambda x: calc_roi(x['annual_cashflow'], x['sale_price']),axis=1)

df.loc[:, "dscr"] = df.apply(lambda x: calc_dscr(x['annual_noi'],x['monthly_repayments']),axis=1)

df.loc[:, "growth"] = calc_expected_growth_rate(7,4)

df.loc[:, "break_even"] = df.apply(lambda x: calc_break_even(x['annual_expenses'], x['monthly_repayments'], x['annual_rent']),axis=1)

df.loc[:, 'score'] = df.apply(lambda x: calc_investment_score(x['cashflow'], x['cap_rate'], x['roi'], x['dscr'], x['growth'], x['break_even']),axis=1)
```
