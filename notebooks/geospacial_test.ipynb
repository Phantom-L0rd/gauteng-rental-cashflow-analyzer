```python
import json

with open("../data/processed/gauteng_adm2.geojson") as f:
    gj = json.load(f)

# List of municipalities and colors
munis = ["City of Johannesburg", "City of Tshwane", "Ekurhuleni", "West Rand", "Sedibeng"]
colors = [25, 0, 0, 0, 0]

# Prepare DataFrame with one row per feature
data = []
for feat in gj["features"]:
    name = feat["properties"]["adm2_name"]
    # data.append({"municipality": name, "dummy": 1})
    color = colors[munis.index(name)] if name in munis else "lightgrey"  # color unmatched features grey
    data.append({"municipality": name, "color": color})

df = pd.DataFrame(data)
df["z"] = [1, 2, 3, 4, 5]
df
```

```python
df["municipality_clean"] = df["municipality"].str.strip().str.lower()
for feat in gj["features"]:
    feat["properties"]["adm2_name_clean"] = feat["properties"]["adm2_name"].strip().lower()

```

```python
import osmnx as ox
import geopandas as gpd

gauteng = ox.geocode_to_gdf("Gauteng, South Africa")

suburbs = ox.features_from_place(
    "Gauteng, South Africa",
    tags={"place": ["suburb", "neighbourhood"]}
)

suburbs = suburbs.to_crs(epsg=4326)
```

```python
suburbs.head()[["name", "place", "geometry"]]
```

```python
suburbs = suburbs[suburbs.geometry.type.isin(["Polygon", "MultiPolygon"])]
suburbs = suburbs[~suburbs["name"].isna()]

suburbs["suburb_name"] = suburbs["name"].str.lower().str.strip()
suburbs = suburbs[["suburb_name", "geometry"]]
suburbs = suburbs.dissolve(by="suburb_name", as_index=False)
```

```python
suburbs
```

```python
output_path = "../data/processed/suburbs.geojson"

suburbs.to_file(
    output_path,
    driver="GeoJSON"
)

print("Saved:", output_path)
```

```python
places_search = ox.features_from_place(
    "Gauteng, South Africa",
    tags={"place": ["city", "town"]}
)

places = places_search.to_crs(epsg=4326)
```

```python
places = places[
    places.geometry.type.isin(["Polygon", "MultiPolygon"])
]

places = places[~places["name"].isna()]

places["place_name"] = (
    places["name"]
    .str.lower()
    .str.strip()
)

places = places[["place_name", "place", "geometry"]]

places = places.dissolve(by="place_name", as_index=False)
```

```python
target_cities = [
    "akasia","alberton","bedfordview","benoni","boksburg","brakpan",
    "bronkhorstspruit","carletonville","centurion","cullinan",
    "edenvale","eikenhof","evaton","fochville","ga-rankuwa",
    "germiston","hammanskraal","heidelberg","johannesburg",
    "katlehong","kempton park","krugersdorp","mabopane","meyerton",
    "midrand","nigel","pretoria","randburg","randfontein","rayton",
    "roodepoort","sandton","sebokeng","soshanguve","soweto","springs",
    "temba","tembisa","vaal marina","vanderbijlpark","vereeniging",
    "walkerville","westonaria","wonderboom"
]

cities = places[places["place_name"].isin(target_cities)]
cities
```
